# class

# Abstraction


Data abstraction means showing only the essential features and hiding the complex internal details. Technically, in Python abstraction is used to hide the implementation details from the user and expose only necessary parts, making the code simpler and easier to interact with.

Real Life Example: In a graphics software, multiple shapes are available such as Circle, Rectangle and Triangle. All shapes share common data and support the same set of actions, but the internal details are hidden:

Common data: name, color, border thickness, fill style.
Common actions: draw(), resize(), getArea().


# Abstract Base Class

In Python, an Abstract Base Class (ABC) is used to achieve data abstraction by defining a common interface for its subclasses. It cannot be instantiated directly and serves as a blueprint for other classes.

Abstract classes are created using abc module and @abstractmethod decorator, allowing developers to enforce method implementation in subclasses while hiding complex internal logic.






why to use abstarct classes 
reasons 
1.ensures that subclasses implement specific methods
2. promotes code consistency and resueability

how to create a abstract class 

```python
from abc import ABC,abstractmethod
class Animal(ABC):
    @abstractmethod
    def make_sound(self):
        pass


```

 here animal is an abstract and make_sound() is an abstarct method that must be defined in child classes 
 

eg of a sub class 

```python
class Dog(Animal):
    def make_sound(self):
        return "bark"
        
```

dog implements make_sound() so it can be instantiated 

if make_sound() is missing , python will raise an error

# property decorator

a @property decorator let`s you define methods that acts like attributes,making access more dynamic 

```python
# ***********************property decorator   "************************
class Student:
   def __init__(self,phy,chem,maths):
      self.phy = phy
      self.chem = chem
      self.maths = maths
   # @property makes it look like a function which changes output each time the values area changed 
   def avg(self):
      return int((self.phy+self.chem+self.maths)/3)
kyo = Student(92,78,99)
print(kyo.avg())
```

usage :- use @property for computed attributes without exposing methods

# polymorphism

polymorphism allows methds in different classes to share the same name but behave completely differently

eg A darw() method in circle and rectangle classes perform shape specific drawing 

types :- 
1. compile - time polymorphism : Achived by method overloading (less commmon in python)
2. run - time polymorphism : Achieved via method overriding in inheritance 

```python
class Animal:
    def speak(self):
        pass
class Dog(Animal):
    def speak(self):
        return "woof"
class Cat(Animal):
    def speak(self):
        return "Meow"
animals = [Dog(),Cat()]
for animal in animals:
    print(animal.speak())
```
